# Data Preprocessing 
- step-1 : Label Encoder(ham=0,spam=1)
- step-2: Text Processing

```python
df['label'] = df['label'].replace({'ham':0,'spam':1})
df.head()
```

```python
wnl = WordNetLemmatizer()
```

```python
def poslemma(pos_tag):
    word,pos=pos_tag
    if pos.startswith('R'):
        pos = 'r'
    elif pos.startswith('V'):
        pos = 'v'
    elif pos.startswith('J'):
        pos ='a'
    else:
        pos='n'
    return wnl.lemmatize(word,pos=pos)

def lematization(sent):
    tokens = sent.split() # spliting the sentence in to words 
    pos = pos_tag(tokens) # taging parts of speatch 
    return " ".join([poslemma(tag) for tag in pos]) 
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
def tokenization(text):
    text = text.split()
    text_list = [word for word in text if word not in stop]
    return " ".join(text_list)
```

```python
def textprocessing(sample):
    sample = sample.lower()
    sample = re.sub(r'[^a-z0-9]'," ",sample)
    sample = lematization(sample)
    return sample
```

```python
df.dropna(axis=0,inplace=True)
df['clear_text'] = df['text'].apply(textprocessing)
df.head()
```

```python
cv = CountVectorizer() # less than 5000
tfidf = TfidfVectorizer() # less then 5000 feature 
hashv = HashingVectorizer() # is prefered is feature are more than 5000 +
```

```python

x_tfidf = tfidf.fit_transform(df['clear_text']).toarray()
#x_hashv = hashv.fit_transform(df['clear_text']).toarray()
```

```python
len(tfidf.get_feature_names())
```

```python
df['label'] = df['label'].astype(int) # converting the label data to integer 
y = df['label'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x_tfidf,y,test_size=0.3)
```

```python
model_nb = MultinomialNB()
model_dt = DecisionTreeClassifier()
```

```python
model_nb.fit(x_train,y_train)
```

```python
# Evaluation 
y_pred = model_nb.predict(x_test)
cm = ConfusionMatrix(y_test,y_pred)
cm
```
