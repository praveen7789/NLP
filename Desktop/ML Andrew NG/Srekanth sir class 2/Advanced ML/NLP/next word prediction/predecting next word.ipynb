```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import names
from nltk import word_tokenize
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
from pandas_ml import ConfusionMatrix
```

```python
with open('./copurs_sample_text.txt','r', encoding='latin-1') as f:
    data = f.read()
data
```

```python
data = data.lower()
data = re.findall(r'[a-z]+',data)
#words = word_tokenize(data)
data
```

```python
data_list = data
```

```python
from nltk import ngrams , bigrams, trigrams
```

```python
bi = list(bigrams(data_list))
```

```python
df = pd.DataFrame(bi)
df.head()
```

```python
df.columns=['first','second']
df.head()
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder() # list of transcation into DF
```

```python
bolean_tr =te.fit_transform(df['second'])
df_1 = pd.DataFrame(data = bolean_tr, columns=te.columns_,index=df['first'])
df_1.head()
```

```python
# Association rules , Apriori
from mlxtend.frequent_patterns import apriori , association_rules
```

```python
df_ap = apriori(df_1,min_support=0.1, use_colnames=True)
df_ap
```

```python
df_ar = association_rules(df_ap,metric='confidence',min_threshold=0.6)
df_ar
```

```python
df_lf = association_rules(df_ap,metric='lift',min_threshold=1.1)
df_lf
```

```python
gp = df.groupby(['first'])
# hed = gp.head()
# hed
gp.head()[gp.head()['first']=='however']
```

```python
hed = gp.head()
hed1 = hed['first'].unique()
len(hed1)
```

```python
gp2=gp.get_group('however')
```

```python
con = gp2.groupby(['second']).count()
con.reset_index(drop=True,inplace=True)
```

```python
gp2=gp.get_group('however')
con = gp2.groupby(['second']).count()
con.reset_index(drop=True,inplace=True)

final = gp2.groupby(['first','second']).count()
final.reset_index(inplace=True)
final['count']=con['first']
```

```python
finaly=pd.DataFrame(columns=['first', 'second', 'count'])
```

```python

for i in hed1:
    #print(i)
    gp2=gp.get_group(i)
    con = gp2.groupby(['second']).count()
    con.reset_index(drop=True,inplace=True)

    final = gp2.groupby(['first','second']).count()
    final.reset_index(inplace=True)
    final['count']=con['first']
    
    finaly = finaly.append(final,ignore_index=True)
```
