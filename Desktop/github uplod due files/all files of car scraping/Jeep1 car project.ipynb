# Web Scraping to extract jeep cars

```python
# requitred libraries
import urllib.request # To request the URL to get the data
from bs4 import BeautifulSoup # it is used to convert the raw data in to structral data
```

```python
html="https://www.carwale.com/jeep-cars/compass/" # official site of jeep
```

```python
html_doc=urllib.request.urlopen(html) 
```

```python
soup=BeautifulSoup(html_doc, 'html.parser') 
print(soup.prettify())
```

```python
span=soup.find_all('span', class_="variant__price")
Price=[]
for i in span:
    Price.append(i.text)
Price    
```

```python
import re
pr = ','.join(Price)
Price = re.findall(r'\d+.\d',pr)
Price
```

```python
len(Price)
```

```python
div=soup.find_all('div', class_='variant__name')
Varient=[]
for i in div:
    
    Varient.append(i.text)
Varient
```

```python
len(Varient)
```

```python
span=soup.find_all('span', class_="variant__specs")
cpm=[]
for sp in span:
    
    cpm.append(sp.text)
len(cpm)
```

```python
len(cpm)
```

```python
CC=[]
for i in range(len(cpm)):
    
    CC.append(cpm[i].split(sep=',')[0])
CC    
    
```

```python
# converting list to string and applying regular expressions
import re
cc = ','.join(CC)
CC = re.findall(r'\d+.\d',cc)
CC
```

```python
len(CC)
```

```python
EngineType=[]
for i in range(len(cpm)):
    
    EngineType.append(cpm[i].split(sep=',')[1])
EngineType
```

```python
len(EngineType)
```

```python
TransmissionType=[]
for i in range(len(cpm)):
    
    TransmissionType.append(cpm[i].split(sep=',')[2])
TransmissionType
```

```python
Milage=[]
for i in range(len(cpm)):
    
    Milage.append(cpm[i].split(sep=',')[3])
Milage
```

```python
# converting list to string and applying regular expressions
import re
Mile = ','.join(Milage)
Mileage = re.findall(r'\d+.\d',Mile)
Mileage
```
