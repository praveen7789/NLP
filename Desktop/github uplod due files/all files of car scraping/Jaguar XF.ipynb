```python
import urllib.request
```

```python
html='https://www.cardekho.com/carmodels/Jaguar/Jaguar_XF'
```

```python
html_doc=urllib.request.urlopen(html)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
```

```python
span=soup.find_all('tr')
cp=[]
for i in range(len(span)):
    
    cp.append(span[i].a)
cp    
```

```python
cp=list(filter(None.__ne__, cp)) # to clean None value this is the code
Varient=[]
for i in range(len(cp)):
    Varient.append(cp[i].text)
Varient=Varient[0:5]   
Varient
    

```

```python
#class="pricevalue"
span=soup.find_all('td', class_="pricevalue")
Pri=[]
for i in span:
    Pri.append(i.text)
Pri    
```

```python
import re
Price=[]

pr = ','.join(Pri)
Pri = re.findall(r'\d+.\d+',pr)
for i in range(len(Pri)):
    Price.append(float(Pri[i])*100000)
Price=Price[0:5] 
Price


    
    
  
```

```python
span=soup.find_all('span', class_="kmpl")
cpm=[]
for sp in span:
    
    cpm.append(sp.text)
cpm=cpm[0:5]
cpm
```

```python
CCc=[]
for i in range(len(cpm)):
    
    CCc.append(cpm[i].split(sep=',')[0])
CCc    
    
```

```python
import re
cc = ','.join(CCc)
CC = re.findall(r'\d+.\d',cc)
CC=[int(i) for i in CC]
CC
```

```python
EngineType=[]
for i in range(len(cpm)):
    
    EngineType.append(cpm[i].split(sep=',')[2])
EngineType
```

```python
TransmissionType=[]
for i in range(len(cpm)):
    
    TransmissionType.append(cpm[i].split(sep=',')[1])
TransmissionType
```

```python
Milage=[]
for i in range(len(cpm)):
    
    Milage.append(cpm[i].split(sep=',')[3])
Milage
```

```python
import re
Mile = ','.join(Milage)
Mileage = re.findall(r'\d+.\d+',Mile)
Mileage=[float(i) for i in Mileage]
Mileage
```

```python
Brand=['Jaguar','Jaguar','Jaguar','Jaguar','Jaguar']
```

```python
ModelName=['Jaguar XF','Jaguar XF','Jaguar XF','Jaguar XF','Jaguar XF']
```

```python
SeatCapacity=[5,5,5,5,5]
```

```python
#import pandas to convert list to data frame
import pandas as pd
df=pd.DataFrame(Brand,columns=['Brand'])
df['ModelName']=ModelName
df['Varient']=Varient
df['EngineType']=EngineType
df['CC']=CC
df['TransmissionType']=TransmissionType
df['SeatCapacity']=SeatCapacity
df['Mileage']=Mileage
df['Price']=Price
df
```
